# Branching Strategies

Branching strategies help manage the workflow of features, bug fixes, and releases

example: GITFLOW

Main Branch (main): main branch contains production-ready code.


Develop Branch (develop): develop branch contains code that is being tested and is almost ready for release.


Feature Branches (feature/*): Feature branches are used to develop new features for the upcoming releases.


Release Branches (release/*): Release branches support preparation of a new production release.


Hotfix Branches (hotfix/*): Hotfix branches are used to quickly fix production issues.

```python
#Initialize Gitflow
git flow init
```

```python
#Creating Feature Branches
git flow feature start new-feature

#Develop the feature and commit your changes
git add .
git commit -m "Add new feature"

#Merge the feature branch into the main branch
git flow feature merge main

#finish the feature branch
git flow feature finish new-feature
```

```python
#Creating Release Branches
git flow release start v1.0

#Finalize and close the release branch
git flow release finish v1.0
```

```python
#Creating Hotfix Branches
git flow hotfix start fix-urgent-bug

#finish the hotfix branch
git flow hotfix finish fix-urgent-bug
```

# Collaborative Development Workflows

Pull Requests

Using Pull requests (PRs) we can review and discuss changes before they are merged into the main codebase.

creating a pull request

```python
#Push Your Branch to Remote
git push origin feature/new-feature

#Create a Pull Request
git pull-request origin main
```

Code Reviews

Code reviews maintain code quality and knowledge sharing. Reviewers can comment on the pull request and request changes if needed. Once the changes are approved, the pull request can be merged

```python
#Fork the Repository
git fork

#Clone the Forked Repository
git clone https://github.com/abhinayr/vms.git
cd vms

#Add the Original Repository as Remote
git remote add origin https://github.com/abhinayr/vms.git

#Sync Fork
git fetch upstream
git checkout main
git merge upstream/main

#Create a New Branch for Your Feature
git checkout -b feature/new-feature

#Make Changes and Commit
git add .
git commit -m "Add new feature"

#Push Your Branch
git push origin feature/new-feature

#Create a Pull Request
git pull-request origin main

#Review the Pull Request
git pull-request review
```
