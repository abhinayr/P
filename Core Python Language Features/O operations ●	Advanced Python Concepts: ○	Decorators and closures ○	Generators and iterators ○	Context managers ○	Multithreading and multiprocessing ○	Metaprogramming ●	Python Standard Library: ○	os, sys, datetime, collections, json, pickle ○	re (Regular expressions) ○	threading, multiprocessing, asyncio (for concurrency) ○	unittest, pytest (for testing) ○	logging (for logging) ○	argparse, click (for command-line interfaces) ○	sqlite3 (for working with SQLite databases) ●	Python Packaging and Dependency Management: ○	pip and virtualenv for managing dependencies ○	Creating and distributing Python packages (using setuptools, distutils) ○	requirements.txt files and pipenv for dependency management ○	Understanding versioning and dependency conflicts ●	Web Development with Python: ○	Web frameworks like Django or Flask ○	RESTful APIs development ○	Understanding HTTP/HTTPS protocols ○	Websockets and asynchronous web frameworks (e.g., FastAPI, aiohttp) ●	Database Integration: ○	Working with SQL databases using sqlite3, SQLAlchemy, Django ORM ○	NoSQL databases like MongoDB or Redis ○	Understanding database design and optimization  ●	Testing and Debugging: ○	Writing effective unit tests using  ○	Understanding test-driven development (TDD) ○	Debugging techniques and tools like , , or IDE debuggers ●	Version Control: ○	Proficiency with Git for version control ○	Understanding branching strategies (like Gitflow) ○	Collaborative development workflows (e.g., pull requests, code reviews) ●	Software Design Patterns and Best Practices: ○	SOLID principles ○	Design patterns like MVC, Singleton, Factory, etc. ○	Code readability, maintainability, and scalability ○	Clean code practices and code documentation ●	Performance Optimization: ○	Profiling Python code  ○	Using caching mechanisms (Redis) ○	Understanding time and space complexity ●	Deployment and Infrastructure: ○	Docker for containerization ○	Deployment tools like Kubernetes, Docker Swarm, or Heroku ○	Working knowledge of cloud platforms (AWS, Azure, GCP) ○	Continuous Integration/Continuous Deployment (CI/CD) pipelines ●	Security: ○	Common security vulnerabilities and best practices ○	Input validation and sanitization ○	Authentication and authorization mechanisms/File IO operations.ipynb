

* **The `open()` function:** It's the key to accessing files in Python. Here's its common syntax:

```
file_object = open('file.txt', 'r')
```

           - `'r'`:  Open for reading (default).

           - `'w'`:  Open for writing (truncates an existing file or creates a new one).

           - `'a'`:  Open for appending (adds to the end of an existing file or creates a new one).

           - `'r+'`: Open for both reading and writing.

           - `'b'`:  Binary mode (for non-text files).


* **The `close()` method:** Always close files when you're finished to release resources.

```
file_object.close()
```

**Examples**

**1. Reading from a File**

```
file = open('file.txt', 'r')
content = file.read()  # Read entire content
print(content)
file.close()

# Reading line by line:
file = open('file.txt', 'r')
for line in file:
    print(line, end='')  # end='' prevents extra newline
file.close()
```

**2. Writing to a File**

```
file = open('file.txt', 'w')
file.write("This is a new line to write.\n")
file.write("And another one.\n")
file.close()
```

**3. Appending to a File**

```
file = open('file.txt', 'a')
file.write("Appended some more text.\n")
file.close()
```

* **`os` module:** Offers more file and directory manipulation functions (renaming, deleting, checking paths, etc.):

```
 import os
if os.path.exists('file.txt'):
    os.remove('file.txt')  # Delete a file
```

**Important Notes**

* **Error Handling:** Use `try-except` blocks to handle potential errors during file operations (e.g., file not found).
* **Resource Management:** Always close files, or use the `with` statement, to avoid file system resource leaks.
