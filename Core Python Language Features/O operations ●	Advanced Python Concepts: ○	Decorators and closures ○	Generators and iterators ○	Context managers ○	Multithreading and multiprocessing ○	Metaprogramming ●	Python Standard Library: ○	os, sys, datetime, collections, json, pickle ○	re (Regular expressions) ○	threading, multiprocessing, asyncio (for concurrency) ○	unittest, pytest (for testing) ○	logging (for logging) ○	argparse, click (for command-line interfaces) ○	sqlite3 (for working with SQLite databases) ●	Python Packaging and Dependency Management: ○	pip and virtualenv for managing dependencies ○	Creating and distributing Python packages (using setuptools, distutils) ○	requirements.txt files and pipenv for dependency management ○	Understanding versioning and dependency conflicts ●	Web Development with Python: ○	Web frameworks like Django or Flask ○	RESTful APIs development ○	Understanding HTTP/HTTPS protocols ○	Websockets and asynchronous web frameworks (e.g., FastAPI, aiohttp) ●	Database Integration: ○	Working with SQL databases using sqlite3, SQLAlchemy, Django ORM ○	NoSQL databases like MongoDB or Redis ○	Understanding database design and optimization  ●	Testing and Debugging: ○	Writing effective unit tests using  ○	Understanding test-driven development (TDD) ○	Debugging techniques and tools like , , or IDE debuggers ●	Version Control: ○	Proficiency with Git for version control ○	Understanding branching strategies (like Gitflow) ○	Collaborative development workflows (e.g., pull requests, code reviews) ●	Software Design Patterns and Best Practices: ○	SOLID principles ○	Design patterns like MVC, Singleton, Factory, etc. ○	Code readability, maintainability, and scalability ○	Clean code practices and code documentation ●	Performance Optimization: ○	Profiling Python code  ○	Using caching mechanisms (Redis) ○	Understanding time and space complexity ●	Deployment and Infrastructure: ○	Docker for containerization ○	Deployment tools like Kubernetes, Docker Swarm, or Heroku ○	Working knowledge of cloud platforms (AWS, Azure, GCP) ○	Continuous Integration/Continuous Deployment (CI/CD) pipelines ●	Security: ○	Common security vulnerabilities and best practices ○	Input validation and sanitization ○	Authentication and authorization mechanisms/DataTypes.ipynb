**Data Types**

1. **Numeric Types**

   * **int (integers):** Whole numbers without decimals.

```
age = 30
print(type(age))
```

* **float (floating-point numbers):** Numbers with decimal points.

```
temp = 24.5
print(type(temp))
```

* **complex (complex numbers):** Numbers with a real and imaginary part (like 2+3j).

```
z = 2 + 5j
print(type(z))
```

2. **Sequence Types**

   * **str (strings):** Ordered sequences of characters.

```
name = "Alice"
print(type(name))
```

* **list:** Ordered, mutable (changeable) collections of items.

```
list = ["eggs", "milk", "bread"]
print(type(list))
```

* **tuple:** Ordered, immutable (unchangeable) collections of items.

```
co = (10, 20)
print(type(co))
```

3. **Boolean**

   * **bool:** Logical values representing True or False.

```
is_raining = True
print(type(is_raining))
```

4. **Mapping Type**

   * **dict (dictionary):** Unordered collections of key-value pairs.

```
student = {"name": "Bob", "age": 25, "city": "New York"}
print(type(student))
```

**Type Conversion (Casting)**

You can convert between data types using the following functions:

* `int(x)`: Converts x to an integer.
* `float(x)`: Converts x to a floating-point number.
* `str(x)`: Converts x to a string.
* `list(x)`:  Converts a sequence (string, tuple, etc.) into a list.
* `tuple(x)`: Converts a sequence into a tuple.

**Example**

```
price = "19.99"
quantity = 5
total_cost = float(price) * quantity
print(type(price))
print(total_cost)
```
